```python
# fill NaNs with the most frequent value from one column.
# to fill every column with its own most frequent value

data=data.apply(lambda x: x.fillna(x.value_counts().index[0]))
print(data.isnull().sum().sum())
print(" ")
print(data.isnull().sum())
```

Making Price Category

```python
# to divide 'Prices_for_2' according to 'Price_Range'
data['Price_Range']= pd.cut(x=data['Price_for_2'], bins=[0,500,1000,1500,2000,3000,5000,9000], labels=['cost < 500', '500- 1000', '1000- 1500', '1500- 2000', '2000- 3000', '3000- 5000', 'cost > 5000'])
```

```python
# sort the 'Dining_Rating' values in ascending order
data.sort_values(by='Dining_Rating', ascending=True)
```

```python
#rating<3.9= Average
#rating>3.9= Good
#rating>4.3= Excellent

data['Dining_Rating_Category']=pd.cut(x=data['Dining_Rating'], bins=[0,3.9,4.3,5], labels=['Average','Good', 'Excellent'])
```

```python
#rating<3.9= Average
#rating>3.9= Good
#rating>4.3= Excellent

data['Delivery_Rating_Category']=pd.cut(x=data['Delivery_Rating'], bins=[0,3.9,4.3,5], labels=['Average','Good', 'Excellent'])
data
```

```python
# grouping of the data according to categories
data2= data.groupby('Restaurant_Name')[['Category','Price_for_2','Price_Range','Dining_Rating_Category','Delivery_Rating_Category']].mean().head()
```

```python
# remove the columns which are not required
data.drop(['Website','Phone_No','Price_for_2','Dining_Rating','Delivery_Rating'], axis=1, inplace=True)
data
```

```python
# make the 'Seating' column in data in uniform case
data['Seating'].str.lower()
data['Seating'].replace("indoor", "Indoor", inplace = True)
data['Seating'].replace("both", "Both", inplace = True)
data['Seating'].replace("outdoor", "Outdoor", inplace = True)
data['Seating'].replace("outoor", "Outdoor", inplace = True)
data['Seating'].replace("Outoor", "Outdoor", inplace = True)
```

```python
# make the 'Serves_Alcohol' column in data in uniform case
data['Serves_Alcohol'].str.lower()
data['Serves_Alcohol'].replace("no", "No", inplace = True)
data['Serves_Alcohol'].replace("both", "Both", inplace = True)
data['Serves_Alcohol'].replace("yes", "Yes", inplace = True)
```
